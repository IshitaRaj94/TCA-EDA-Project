```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Customer Churn.csv')
df
```

#replacing blanks with 0 as tenure is 0 and no total charges are recorded

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ","0")
df["TotalCharges"]=df["TotalCharges"].astype("float")
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

# converted 0 and 1 values of senior citizen to yes/no to make it easier to understand

```python
def conv(value):
    if value==1:
        return "yes"
    else:
        return "no"

df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

```python
df.head(30)
```

```python
ax=sns.countplot(x='Churn' ,data=df)

ax.bar_label(ax.containers[0])
plt.title("Count of customers by Churn")
plt.show()
```

```python
plt.figure(figsize=(3,4))
gb=df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'],labels=gb.index,autopct =" %1.2f%%")
plt.title("Percentage of Churned Customers")
plt.show()
```

# from the pie-chart we can see that 26.54% of the customers have churned out

```python
plt.figure(figsize=(4,4))
sns.countplot(x= "gender", data=df,hue="Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x= "SeniorCitizen", data=df,hue="Churn")
plt.title("Churn by Senior Citizen")
plt.show()
```

# comparativley a greater % of senior citizens have churned out

```python
counts = (
    df
    .groupby(['SeniorCitizen', 'Churn'])
    .size()                           # raw counts
    .unstack(fill_value=0)            # columns = Churn values
)

# 2️⃣   Convert each SeniorCitizen row to percent of that row’s total
percent = counts.div(counts.sum(axis=1), axis=0) * 100

# 3️⃣   Plot stacked bar chart
ax = percent.plot(kind='bar',
                  stacked=True,
                  figsize=(5,4),
                  colormap='Set2',
                  edgecolor='none')

# 4️⃣   Add % labels in the middle of each bar segment
for p in ax.patches:
    h = p.get_height()            # segment height in %
    if h > 0:                     # skip 0‑height slices
        ax.text(
            p.get_x() + p.get_width()/2,
            p.get_y() + h/2,
            f'{h:.1f}%',
            ha='center', va='center',
            fontsize=9, color='black'
        )

# 5️⃣   Final cosmetic touches
ax.set_xlabel('Senior Citizen (0 = No, 1 = Yes)')
ax.set_ylabel('Percent of Senior‑Citizen group')
ax.set_title('Churn by Senior‑Citizen Status (% of Group)')
ax.legend(title='Churn', frameon=False)
plt.tight_layout()
plt.show()
```

```python
cat_cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# --- subplot grid size ---
n_cols = 3
n_rows = int(np.ceil(len(cat_cols) / n_cols))

fig, axes = plt.subplots(n_rows, n_cols,
                         figsize=(5 * n_cols, 4 * n_rows),
                         sharey=False)        # separate y‑axes look cleaner

axes = axes.flatten()                         # easy 1‑D iteration

for i, col in enumerate(cat_cols):
    sns.countplot(
        data=df, x=col,
        ax=axes[i],
        order=df[col].value_counts().index,   # optional: consistent order
        palette="Set2" ,                       # pick a palette you like
     hue="Churn"                        # uncomment if you want stacked hue
    )
    axes[i].set_title(f"Count of {col}")
    axes[i].set_xlabel("")                    # keep tidy; adjust if you prefer
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis="x", rotation=45)

# remove any empty subplots (if len(cat_cols) < n_rows*n_cols)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

fig.tight_layout()
plt.show()
```
